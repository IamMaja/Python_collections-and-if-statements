### # Sets use curley bracks

```python
regions = {"Gauteng", "Limpopo", "Gauteng"}
print(regions)
```

### Set function can be used if you forgot the type of brackets sets require

```python
regions = {"Gauteng", "Limpopo", "Gauteng"}
regions = set(("Gauteng", "Limpopo", "Gauteng"))
```

### Adding item to the set

```python
regions.add("Western Cape")
print(regions)
```

```python
reg2 = {"Free State", "Eastern Cape"}
print(reg2)
```

### combining the sets using union funtion

```python
final_reg = regions.union(reg2)
print(final_reg)
```

python checks from top to bottom
1. is the score >= 80
2. if not, is the score >= 60
3. if not, is score >= 50
4. Else -> Fail

```python
score = 75
if score >= 80: #ranges 80 - 100
    print("Distinction")
elif score >= 60: #ranges 60 - 80
    print("Merit")
elif score >= 50:#ranges 50 - 60
    print("Pass")
else:#ranges 0 - 50
    print("Fail")
```
